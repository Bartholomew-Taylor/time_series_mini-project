# Baseline Established 
## Rolling average with a 365 period used
## Visualization Created

```python
#creates a DataFrame for our comparisons
eval_df = pd.DataFrame(columns=['model_type', 'target_var', 'rmse'])
```

```python
#creates our baseline prediction
period=365
rolling_temp = round(train['AverageTemperature'].rolling(period).mean()[-1],2)
```

```python
#establishes variable for prediction as our baseline
yhat_df = modeling.make_baseline_predictions(rolling_temp, validate)
```

## RMSE 5 is the number to beat

```python
#plots the baseline over train and validate

for col in train.columns:
    modeling.plot_and_eval(col, train, validate, yhat_df)
```

```python
#adds basline to our comparison dataframe
for col in train.columns:
    eval_df = modeling.append_eval_df('moving_avg', col, validate, yhat_df, eval_df)
eval_df
```

# Modeling
## First Model Generated: Holt Linear

```python
#creates and fits the first model

col = 'AverageTemperature' 
model = Holt(train[col], exponential=False, damped=False)
model = model.fit(optimized=True)
yhat_temp = model.predict(start = validate.index[0], end = validate.index[-1])
```

```python
#generates the yhat values 




for col in train.columns:
    model = Holt(train[col], exponential=False, damped=True)
    model = model.fit(optimized=True)
    yhat_values = model.predict(start = validate.index[0], end = validate.index[-1])
    yhat_df[col] = round(yhat_values, 2)
        
```

## Model performed at an RMSE of 20

```python
#plots the predictions of the first model

for col in train.columns:
    modeling.plot_and_eval(col, train, validate, yhat_df)
```

```python
#adds our first model to the comparison dataframe

eval_df = modeling.append_eval_df('holts_optimized', col, validate, yhat_df, eval_df)
```

```python
eval_df
```

## Second model generated was Holts Seasonal Trend 

```python
#fits the second model

hst_temp = ExponentialSmoothing(train.AverageTemperature, seasonal_periods=480, 
                                trend='add', seasonal='add', damped = True).fit()

```

```python
#generates the predictions for the second model

yhat_df = pd.DataFrame({'AverageTemperature': hst_temp.forecast(validate.shape[0])},
                          index=validate.index)
```

## Model perfomed with an RMSE of 2

```python
#plots the second model against the train and validate data

for col in train.columns:
    modeling.plot_and_eval(col, train, validate, yhat_df)
```

```python
#adds the second model to the comparison dataframe

eval_df = modeling.append_eval_df('holts_seasonal', col, validate, yhat_df, eval_df)
```

```python
eval_df
```

```python
#resplits the data to ensure that the data is ready for the test phase

train, validate, test = wrangle.split_time(df)
```

```python
#generates the test predictions

yhat_df = pd.DataFrame({'AverageTemperature': hst_temp.forecast(validate.shape[0] + test.shape[0])})
yhat_df=yhat_df['1959-11-01':]
rmse_temp_total = sqrt(mean_squared_error(test['AverageTemperature'], yhat_df['AverageTemperature']))
```

## Holts Seasonal Trend Model was utilized on Test Data and had a RMSE of 1.65

```python
#plots the test phase of the model with Holts Seasonal Trend model


print('FINAL PERFORMANCE OF MODEL ON TEST DATA')
print('rmse-temp total: ', rmse_temp_total)
for col in train.columns:
    modeling.final_plot(col, train, validate, test, yhat_df)
```
